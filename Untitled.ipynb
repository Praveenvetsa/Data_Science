# TYPE CASTING OR TYPE CONVERSION
int() -- float() -- complex() -- bool() -- str()

```python
# int -> converting another datatype into int datatype
# type casting from other data type to int & float possible except complex & string
# we take only one arugument
print(int(10.123))  # float to int
print(int(True))     # Bool to int
print(int(False))   # Boll to int
print(int('10'))    # string to int
```

```python
int(10+20j) # cannot convert from complex to int
```

```python
int('ten') # cannot convert string to int when we use the text data
```

```python
# converting other datatype to float datatype
print(float(10))      # int to float
print(float(False))     # boolean to float
print(float('11'))    # string to float
```

```python
float(10+20j)  # cannot convert complex to float
```

```python
float('ten') # you cannot convert string datatype to flaot datatype while using the text data
```

```python
# converting other datatype to complex datatype
# --- this only for 1 argument ------
a = complex(10)     # int to complex
b = complex(10.5)  # float to complex
c = complex(True)  # Bool to complex
d = complex(False) # Bool to complex 
e = complex('10')
```

```python
print(a,b,c,d,e)
```

```python
# ----- Now we will check for 2 arguments ------
p1 = complex(10,20)   # int to complex
p2 = complex(10,20.5) # float to complex
p3 = complex('10')    # string to complex, you cannot assign 2 argument
```

```python
print(p1,p2,p3)
```
